```python
# ====================== IMPORTS ======================
import tensorflow as tf
import numpy as np
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# ====================== LOAD DATA ======================
# The previous URLs resulted in a 404 error. Updating to known working URLs for the UCI HAR Dataset.
# For X_train.txt
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00240/UCI%20HAR%20Dataset.zip"
import requests
import zipfile
import io

r = requests.get(url)
z = zipfile.ZipFile(io.BytesIO(r.content))
z.extractall('./UCI HAR Dataset/')

data = pd.read_csv('./UCI HAR Dataset/UCI HAR Dataset/train/X_train.txt', sep=r'\s+', header=None)

# For y_train.txt
labels = pd.read_csv('./UCI HAR Dataset/UCI HAR Dataset/train/y_train.txt', sep=r'\s+', header=None)

# ====================== PREPROCESS ======================
scaler = StandardScaler()
data = scaler.fit_transform(data)

data = data.reshape((data.shape[0], 1, data.shape[1]))

# Adjust labels to be 0-indexed for sparse_categorical_crossentropy
labels = labels - 1

X_train, X_test, y_train, y_test = train_test_split(data, labels, test_size=0.2)

# ====================== MODEL ======================
model = Sequential()
model.add(LSTM(64, input_shape=(1,561), return_sequences=False))
model.add(Dropout(0.3))
model.add(Dense(32, activation='relu'))
model.add(Dense(6, activation='softmax'))

model.compile(loss='sparse_categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

# ====================== TRAIN ======================
model.fit(X_train, y_train, epochs=5, batch_size=32)

# ====================== EVALUATE ======================
loss, acc = model.evaluate(X_test, y_test)
print("Accuracy:", acc)
```
